### Funcion para incorporación de excedentes
[Calcula los ingresos mensuales por usuario (resta el límite del paquete gratuito del número total de llamadas, mensajes de texto y datos; multiplica el resultado por el valor del plan de llamadas; añade la tarifa mensual en función del plan de llamadas). Nota: Dadas las condiciones del plan, ¡esto podría no ser tan trivial como un par de líneas! Así que no pasa nada si dedicas algo de tiempo a ello.]

```python
# 1. Calcular consumo extra
def calc_extra(used, included):
    """Calcula consumo extra si se excede el límite."""
    return max(used - included, 0)

# 2. Calcular costo por unidad
def calc_cost(extra_units, unit_price):
    """Calcula el costo de unidades extra."""
    return extra_units * unit_price

# 3. Convertir MB a GB
def mb_to_gb(mb):
    return mb / 1000

#  Función principal (independiente del DataFrame)
def calculate_monthly_income(
    minutes_used, minutes_included, usd_per_minute,
    messages_sent, messages_included, usd_per_message,
    mb_used, mb_included, usd_per_gb,
    monthly_fee):

    # Minutos
    extra_minutes = calc_extra(minutes_used, minutes_included)
    cost_minutes = calc_cost(extra_minutes, usd_per_minute)

    # Mensajes
    extra_messages = calc_extra(messages_sent, messages_included)
    cost_messages = calc_cost(extra_messages, usd_per_message)

    # Datos
    extra_mb = calc_extra(mb_used, mb_included)
    cost_data = mb_to_gb(extra_mb) * usd_per_gb

    return cost_minutes + cost_messages + cost_data + monthly_fee

# 4. Adaptador para cualquier DataFrame
# Esta función mapea columnas → lógica, y es la única que conoce el DataFrame.
def calc_income_from_row(row, col_map):
    return calculate_monthly_income(
        minutes_used=row[col_map['minutes_used']],
        minutes_included=row[col_map['minutes_included']],
        usd_per_minute=row[col_map['usd_per_minute']],

        messages_sent=row[col_map['messages_sent']],
        messages_included=row[col_map['messages_included']],
        usd_per_message=row[col_map['usd_per_message']],

        mb_used=row[col_map['mb_used']],
        mb_included=row[col_map['mb_included']],
        usd_per_gb=row[col_map['usd_per_gb']],

        monthly_fee=row[col_map['monthly_fee']])

# 5. Uso con TU DataFrame actual
# Define el mapa de columnas (esto permite usar cualquier DataFrame):
COLUMN_MAP = {
    'minutes_used': 'minutes_used',
    'minutes_included': 'minutes_included',
    'usd_per_minute': 'usd_per_minute',

    'messages_sent': 'messages_sent',
    'messages_included': 'messages_included',
    'usd_per_message': 'usd_per_message',

    'mb_used': 'mb_used',
    'mb_included': 'mb_per_month_included',
    'usd_per_gb': 'usd_per_gb',

    'monthly_fee': 'usd_monthly_pay'}

# Incoporar al DataFrame actual:
df_total_consumption['total_income'] = df_total_consumption.apply(
    lambda row: calc_income_from_row(row, COLUMN_MAP),
    axis=1)

df_total_consumption
```

#### Calcula el consumo y los ingresos por número de usuarios activos por plan y mes para obtener la rentabilidad real promedio por usuario

```python
# Filtra usuarios activos
df_total_consumption['is_active'] = (
    (df_total_consumption['minutes_used'] > 0) |
    (df_total_consumption['messages_sent'] > 0) |
    (df_total_consumption['mb_used'] > 0)
)

# Agregacion por plan y mes
monthly_plan_summary = (
    df_total_consumption[df_total_consumption['is_active']]
    .groupby(['plan_name', 'month'])
    .agg(
        active_users=('user_id', 'nunique'),
        total_income=('total_income', 'sum'),
        total_minutes=('minutes_used', 'sum'),
        total_messages=('messages_sent', 'sum'),
        total_mb=('mb_used', 'sum')
    )
    .reset_index()
)

# Calcula la rentabilidad promedio por usuario
monthly_plan_summary['arpu'] = (
    monthly_plan_summary['total_income'] /
    monthly_plan_summary['active_users']
)
```

```python
# Comparar la evolución del ARPU (Ingreso promedio por usuario activo) por plan en el tiempo.

import matplotlib.pyplot as plt

for plan in monthly_plan_summary['plan_name'].unique():
    data = monthly_plan_summary[monthly_plan_summary['plan_name'] == plan]
    plt.plot(data['month'], data['arpu'], label=plan)

plt.xlabel('Mes')
plt.ylabel('ARPU (Ingreso promedio por usuario)')
plt.title('Evolución del ARPU por plan')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

- Plan Surf
- Este plan no es el más rentable por usuario, pero presenta un crecimiento sostenido del ARPU a lo largo del tiempo Muestra mayor volatilidad, consistente con un esquema tipo prepago. La tendencia sugiere que los cargos por excedentes están aumentando, lo cual impulsa su rentabilidad.
- Surf mejora su rentabilidad gracias al comportamiento de sobreconsumo, no por su cuota base.

- Plan Ultimate
- Es el plan con mayor ARPU, es decir, más rentable por usuario su ingreso es estable y predecible, el crecimiento es leve, lo que sugiere:
    - Usuarios con consumo controlado
    - Menor dependencia de cargos por exceso
- Ultimate prioriza estabilidad sobre crecimiento acelerado.
